```python
# Importar as bibliotecas
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

### **Conceitos básicos utilizando Grafos**

```python
G_Igrafo = nx.DiGraph()
```

```python
# Criando nossa rede de microsserviços


G_Igrafo.add_edge("microsservico1","microsservico2")
G_Igrafo.add_edge("microsservico1","microsservico10")
G_Igrafo.add_edge("microsservico2","microsservico20")
G_Igrafo.add_edge("microsservico3","microsservico20")
G_Igrafo.add_edge("microsservico4","microsservico20")
G_Igrafo.add_edge("microsservico5","microsservico20")
G_Igrafo.add_edge("microsservico6","microsservico20")
G_Igrafo.add_edge("microsservico7","microsservico20")
G_Igrafo.add_edge("microsservico8","microsservico2")
G_Igrafo.add_edge("microsservico9","microsservico2")
G_Igrafo.add_edge("microsservico10","microsservico3")
G_Igrafo.add_edge("microsservico11","microsservico12")
G_Igrafo.add_edge("microsservico12","microsservico13")
G_Igrafo.add_edge("microsservico13","microsservico14")
G_Igrafo.add_edge("microsservico14","microsservico4")
G_Igrafo.add_edge("microsservico15","microsservico5")
G_Igrafo.add_edge("microsservico16","microsservico5")
G_Igrafo.add_edge("microsservico17","microsservico6")
G_Igrafo.add_edge("microsservico18","microsservico8")
G_Igrafo.add_edge("microsservico19","microsservico9")
G_Igrafo.add_edge("microsservico20","microsservico2")

```

```python
# Mostrar informações da nossa rede
print(nx.info(G_Igrafo))
```

```python
np.random.seed(12)
position = nx.spring_layout(G_Igrafo)
plt.figure(figsize=(20,20))

nx.draw_networkx(G_Igrafo,pos=position,  cmap=plt.get_cmap("prism"),
                 node_size=3000, with_labels = True,arrowsize=10,edge_color='orange',width=1 )
```
